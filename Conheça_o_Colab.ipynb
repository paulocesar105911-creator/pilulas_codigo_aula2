Nota Prevista

```
from sklearn.linear_model import LinearRegression
Horas = [[2],[4],[6],[8]]

Notas = [5,6,8,9]
modelo = LinearRegression()
modelo.fit(Horas, Notas)
previsao = modelo.predict([[2]])
print(f"Nota prevista: {previsao[0]}")
```

Classificação - aprovado/reprovado

```
from sklearn.tree import DecisionTreeClassifier # Corrigido: 'DefisionTree.Classifier' para 'DecisionTreeClassifier'
Nota_freq = [[7,80], [6,60], [9,90], [4,50]] # Corrigido: Assumindo que 9.90 deveria ser [9,90]
Resultado = [1,0,1,0] # Corrigido: Assumindo que '10' era um erro e deveria ser uma classe binária (1 ou 0)

modelo = DecisionTreeClassifier ()
modelo.fit(Nota_freq, Resultado) # Corrigido: 'fit' é um método para treinar o modelo
prediction = modelo.predict ([[6,60]]) # O método 'predict' é usado para fazer previsões após o treinamento

print("Aprovado" if prediction[0] == 1 else "Reprovado") # Corrigido: Parêntese em 'resultado[0]' e usando a variável 'prediction'
```

Credito aprovado/reprovado

```
from sklearn.tree import DecisionTreeClassifier

Clientes = [[3000,500], [2000,1500], [5000,1000], [1500,1200]]
Resultado = [1,0,1,0]

Modelo = DecisionTreeClassifier()
Modelo.fit(Clientes, Resultado)
Previsao = Modelo.predict([[7000,800]])
print("Crédito Aprovado" if Previsao[0] == 1 else "Crédito Negado")
```
